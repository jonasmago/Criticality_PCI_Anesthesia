```python
import os
import pickle
import matplotlib.pyplot as plt
import glob

paths = glob.glob('results/details/*')
path = paths[4]


# Load the .pkl file
with open(path, 'rb') as f:
    data = pickle.load(f)

# Extract the relevant parts
x_z = data.get('part_hist_x')
y_z = data.get('part_hist_y')
x_raw = data.get('part_hist_x_raw')
y_raw = data.get('part_hist_y_raw')

# Plotting
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=False)

axes[0].bar(x_z[:-1], y_z, width=(x_z[1]-x_z[0]), color='skyblue', edgecolor='k')
axes[0].set_title('Z-scored EEG Amplitude Distribution')
axes[0].set_xlabel('Z-score')
axes[0].set_ylabel('Count')

axes[1].bar(x_raw[:-1], y_raw, width=(x_raw[1]-x_raw[0]), color='salmon', edgecolor='k')
axes[1].set_title('Raw Filtered EEG Amplitude Distribution')
axes[1].set_xlabel('Amplitude (µV)')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
